```python
import os
import glob

def delete_images_with_copy(directory):
    # Construct the search pattern for files containing 'copy' in their name
    search_pattern = os.path.join(directory, '*copy*')

    # Iterate over all files matching the search pattern
    for file_path in glob.glob(search_pattern):
        try:
            os.remove(file_path)
            print(f"Deleted: {file_path}")
        except Exception as e:
            print(f"Error deleting {file_path}: {e}")

# Specify the directory containing the images
image_directory = 'smallerrefined/degree_0'

# Call the function to delete images with 'copy' in their filenames
delete_images_with_copy(image_directory)
```
